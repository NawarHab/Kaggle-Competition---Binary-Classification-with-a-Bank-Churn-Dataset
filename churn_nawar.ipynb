**Feature Engineering**
1. Removed unique value features
2. Transformed text feature to numbers features
3. Scaling 0 to 1

```python
df_id = df_train['id']
df_train = df_train.drop(['id','CustomerId', 'Surname'], axis=1)
df_train['Gender'] = df_train['Gender'].replace({'Male': 1, 'Female': 2})
df_train['Geography'] = df_train['Geography'].replace({'France': 1, 'Spain': 2, 'Germany': 3})
# df_train = normalize(df_train, axis=0, norm='l1')
scaler = MinMaxScaler()
df_train = pd.DataFrame(scaler.fit_transform(df_train), columns=df_train.columns)

df_test = df_test.drop(['id','CustomerId', 'Surname'], axis=1)
df_test['Gender'] = df_test['Gender'].replace({'Male': 1, 'Female': 2})
df_test['Geography'] = df_test['Geography'].replace({'France': 1, 'Spain': 2, 'Germany': 3})
# df_train = normalize(df_train, axis=0, norm='l1')
scaler = MinMaxScaler()
df_test = pd.DataFrame(scaler.fit_transform(df_test), columns=df_test.columns)


```

**Neural Network using PyTorch**

```python
import torch
from torch import nn
from torch.utils.data import DataLoader
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python

X_train = df_train.drop('Exited', axis=1)
y_train = df_train[['Exited']]


#print(type(y_train))

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_train, y_train, test_size=0.2, random_state=42)



# Convert to PyTorch tensors
X_train = torch.tensor(X_train.to_numpy(), dtype=torch.float32)
y_train = torch.tensor(y_train.to_numpy(), dtype=torch.float32)
X_test = torch.tensor(X_test.to_numpy(), dtype=torch.float32)
y_test = torch.tensor(y_test.to_numpy(), dtype=torch.float32)
```

```python
# Define the model
class BinaryClassifier(nn.Module):
    def __init__(self):
        super(BinaryClassifier, self).__init__()
        self.layer = nn.Sequential(
            nn.Linear(X_train.shape[1], 100),
            nn.ReLU(),
            nn.Linear(100, 50),
            nn.ReLU(),
            nn.Linear(50, 1),
            nn.Sigmoid()
        )

    def forward(self, x):
        return self.layer(x)
```

```python
# Initialize the model, loss function, and optimizer
model = BinaryClassifier()
criterion = nn.BCELoss()
# criterion = nn.BCEWithLogitsLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
```

```python
# Train the model
for epoch in range(500):
    # Forward pass
    outputs = model(X_train)
    loss = criterion(outputs, y_train)

    # Backward pass and optimization
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    print(f'Epoch {epoch+1}/{500}, Loss: {loss.item()}')
```

```python
# Evaluate the model
model.eval()
with torch.no_grad():
    predictions = model(X_test)
    predictions_r = predictions.round()
    accuracy = (predictions_r == y_test).sum().item() / y_test.size(0)

print(f'Accuracy: {accuracy * 100}%')
```
